Video: DataFrame - Índices Multiníveis

```python
# Cria duas listas que servirão como níveis do índice
outside = ['G1', 'G1', 'G1', 'G2', 'G2', 'G2']
inside = [1, 2, 3, 1, 2, 3]

# Une as listas em pares (tuplas): [('G1', 1), ('G1', 2)...]
hier_index = list(zip(outside, inside))

# Transforma essas tuplas em um objeto MultiIndex oficial do Pandas
hier_index = pd.MultiIndex.from_tuples(hier_index)
```

```python
# Exibe 'outside'
outside
```

```python
# Exibe 'inside'
inside
```

```python
# Combina as listas 'outside' e 'inside' em uma lista de tuplas
# Resultado: [('G1', 1), ('G1', 2), ('G1', 3), ('G2', 1), ('G2', 2), ('G2', 3)
hier_index = list(zip(outside, inside))
```

```python
# Exibe 'hier_index'
hier_index
```

```python
# Converte a lista de tuplas em um objeto MultiIndex oficial do Pandas
# Isso permite que o DataFrame entenda a hierarquia entre os níveis 'G1/G2' e '1/2/3'
hier_index = pd.MultiIndex.from_tuples(hier_index)
```

```python
# Exibe 'hier_index'
hier_index
```

```python
# Cria um DataFrame de 6 linhas e 2 colunas com números aleatórios
# index=hier_index: Aplica os níveis G1/G2 e 1/2/3 como rótulos das linhas
# columns=['A', 'B']: Define os nomes das colunas
df = pd.DataFrame(np.random.randn(6, 2), index=hier_index, columns=['A', 'B'])
```

```python
# Exibe 'df'
df
```

```python
# Primeiro seleciona o nível externo 'G1' (retorna um DataFrame menor)
# E depois seleciona a linha '1' dentro desse grupo (retorna uma Series)
df.loc['G1'].loc[1]
```

```python
# Atribui nomes aos níveis do MultiIndex
# O primeiro nível (G1, G2) passa a se chamar 'Grupo'
# O segundo nível (1, 2, 3) passa a se chamar 'Numero'
df.index.names = ['Grupo', 'Numero']
```

```python
# Exibe 'df'
df
```

```python
# Extrai uma seção transversal do DataFrame
# Pega todas as linhas onde o nível 'Numero' é igual a 1, 
# atravessando todos os grupos do nível superior ('G1', 'G2', etc.)
df.xs(1, level='Numero')
```

Video: Dados Ausentes

```python
# Cria um dicionário onde:
# 'A' tem um valor faltante
# 'B' tem dois valores faltantes
# 'C' está completo
d = {'A':[1, 2, np.nan], 'B':[5, np.nan, np.nan], 'C':[1, 2, 3]}
```

```python
# Exibe 'd'
d
```

```python
# Converte o dicionário 'd' em um DataFrame
# As chaves ('A', 'B', 'C') viram colunas e as listas viram as linhas
df = pd.DataFrame(d)
```

```python
# Exibe 'df'
df
```

```python
# Remove todas as linhas que possuem pelo menos um valor nulo (NaN)
df.dropna()
```

```python
# Mantém apenas as linhas que possuem, no mínimo, 2 valores NÃO-NULOS
df.dropna(thresh=2)
```

```python
# Substitui todos os valores nulos (NaN) pela string 'fill na'
df.fillna(value='fill na')
```

```python
# Preenche os valores nulos da coluna 'A' especificamente
# O valor usado para preencher é a média aritmética dos valores existentes na coluna 'A'
df['A'].fillna(value=df['A'].mean())
```

```python
# Preenche os valores nulos (NaN) com o último valor válido encontrado anteriormente
# Ele "propaga" o dado de cima para baixo
df.ffill()
```

Video: GroupBy

```python
# Cria um dicionário com informações de vendas por empresa e funcionário
data = {'Empresa':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Nome':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Venda':[200,120,340,124,243,350]}
```

```python
# Transforma o dicionário 'data' em um DataFrame estruturado
df = pd.DataFrame(data)
```

```python
# Exibe 'df'
df
```

```python
# Cria um objeto GroupBy, agrupando os dados pela coluna 'Empresa'
# O Pandas agora sabe que 'GOOG', 'MSFT' e 'FB' são categorias distintas
group = df.groupby('Empresa')
```

```python
# Soma todos os valores das colunas numéricas para cada grupo (Empresa)
# O parâmetro numeric_only=True garante que o Pandas não tente somar 
# a coluna 'Nome' (que contém texto)
group.sum(numeric_only=True)
```

```python
# Calcula a média das colunas numéricas (Venda) para cada Empresa
# numeric_only=True: Ignora a coluna 'Nome' pois não se tira média de texto
group.mean(numeric_only=True)
```

```python
# Gera um resumo estatístico detalhado para cada grupo (Empresa)
# Inclui: contagem, média, desvio padrão, valor mínimo, quartis e valor máximo
group.describe()
```

```python
# Agora o Pandas organiza os baldes baseados no nome do funcionário
group = df.groupby('Nome')
```

```python
# Soma os valores numéricos agrupados por 'Nome'
# Como cada nome é único, o resultado mostra a venda individual de cada um,
# mas agora usando o Nome como o índice da tabela.
group.sum(numeric_only=True)
```

```python
# 1. Agrupa por 'Nome'
# 2. Soma os valores (numeric_only=True)
# 3. Localiza (.loc) a linha onde o índice é 'Amy'
group.sum(numeric_only=True).loc['Amy']
```

Video: Concatenar, Juntar e Mesclar

```python
# Cria o primeiro DataFrame (df1) com 4 colunas (A, B, C, D) e 4 linhas
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
```

```python
# Cria o segundo DataFrame (df2)
# Mantém as mesmas colunas (A, B, C, D) para garantir o alinhamento
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7]) 
```

```python
# Cria o terceiro DataFrame (df3)
# Mantém as mesmas colunas (A, B, C, D) para garantir o alinhamento
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
```

```python
# Exibe 'df1'
df1
```

```python
# Exibe 'df2'
df2
```

```python
# Exibe 'df3'
df3
```

```python
# Concatena (cola) os DataFrames df1, df2 e df3 um embaixo do outro
# O resultado é um único DataFrame com todas as 12 linhas
pd.concat([df1,df2,df3])
```

```python
# Concatena os DataFrames horizontalmente (adicionando colunas)
# O Pandas tenta alinhar as linhas que possuem o mesmo índice
pd.concat([df1,df2,df3],axis=1)
```

```python
# DataFrame da esquerda: contém as colunas A, B e a chave de ligação
esquerda = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

# DataFrame da direita: contém as colunas C, D e a mesma chave de ligação
direita = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
# Exibe 'esquerda'
esquerda
```

```python
# Exibe 'direita'
direita
```

```python
# Realiza o merge (união) focando apenas no que é comum aos dois lados
# esquerda: O DataFrame base
# direita: O DataFrame que será acoplado
# how='inner': Mantém apenas as chaves que dão "match" em ambos
# on='key': A coluna que serve de ponte entre as tabelas
pd.merge(esquerda,direita,how='inner',on='key')
```

```python
# DataFrames com chaves compostas
esquerda = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
direita = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
# Realiza a união baseada na combinação obrigatória de key1 e key2
pd.merge(esquerda, direita, on=['key1', 'key2'])
```

```python
# Realiza um Outer Join (União Externa)
# Mantém todas as chaves de ambos os DataFrames
pd.merge(esquerda, direita, how='outer', on=['key1', 'key2'])
```

```python
# Realiza um Right Join (União à Direita)
# Preserva todas as combinações de chaves do DataFrame 'direita'
pd.merge(esquerda, direita, how='right', on=['key1', 'key2'])
```

```python
# Realiza um Left Join (União à Esquerda)
# Preserva todas as linhas do DataFrame 'esquerda'
pd.merge(esquerda, direita, how='left', on=['key1', 'key2'])
```

```python
# DataFrame da esquerda: Chaves K0, K1, K2 no índice
esquerda = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

# DataFrame da direita: Chaves K0, K2, K3 no índice
direita = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
# Exibe 'esquerda'
esquerda
```

```python
# Exibe 'direita'
direita
```

```python
# O join utiliza os índices ('K0', 'K1', 'K2') por padrão
esquerda.join(direita)
```

```python
# Realiza o Outer Join usando os índices
# Resultado: Uma tabela que inclui K0, K1, K2 e K3
esquerda.join(direita, how='outer')
```

```python
# Criando um DataFrame de exemplo com números e strings
df = pd.DataFrame({'col1':[1,2,3,4],'col2':[444,555,666,444],'col3':['abc','def','ghi','xyz']})

# Exibe as 5 primeiras linhas
df.head()
```

```python
# Retorna o número de itens exclusivos (444, 555, 666)
df['col2'].nunique()
```

```python
# Conta quantas vezes cada valor aparece na coluna
df['col2'].value_counts()
```

```python
# Filtra o DataFrame usando duas condições:
# 1. col1 deve ser maior que 2 (Ex: 3, 4...)
# 2. col2 deve ser exatamente 444
# O operador '&' garante que AMBAS as condições sejam verdadeiras.
df[(df['col1'] > 2) & (df['col2'] == 444)]
```

```python
# Cria um dicionario
data = {'A':['foo','foo','foo','bar','bar','bar'],
     'B':['one','one','two','two','one','one'],
       'C':['x','y','x','y','x','y'],
       'D':[1,3,2,5,4,1]}

# Converte o dicionario em um dataframe
df = pd.DataFrame(data)
```

```python
# Exibe 'df'
df
```

```python
# Reestrutura o DataFrame:
# Linhas (Index): Agrupa por 'A' e 'B', criando uma hierarquia de categorias.
# Colunas (Columns): Transpõe os valores únicos de 'C' para o cabeçalho.
# Valores (Values): Extrai dados de 'D' para preencher as células.
df.pivot_table(values='D',index=['A', 'B'],columns=['C'])
```
