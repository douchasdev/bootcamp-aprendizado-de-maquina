Video: Dados Ausentes

```python
# Crio um dicionário
d = {'A':[1, 2, np.nan], 'B':[5, np.nan, np.nan], 'C':[1, 2, 3]}
```

```python
# Exibo 'd'
d
```

```python
# Converto o dicionário 'd' em um dataframe
df = pd.DataFrame(d)
```

```python
# Exibo 'df'
df
```

```python
# Removo as linhas que possuem pelo menos um dado nulo (NaN)
df.dropna()
```

```python
# Mantenho apenas as linhas que possuem, no mínimo, 2 dados NÃO-NULOS
df.dropna(thresh=2)
```

```python
# Substituo todos os dados nulos pela string 'fill na'
df.fillna(value='fill na')
```

```python
# Preencho os dados nulos da coluna 'A' com a média aritmética de seus dados não nulos
df['A'].fillna(value=df['A'].mean())
```

```python
# Preencho os dados nulos com o último dado não nulo encontrado
df.ffill()
```

Video: GroupBy

```python
# Crio um dicionário
data = {'Empresa':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Nome':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Venda':[200,120,340,124,243,350]}
```

```python
# Converto o dicionário 'data' em um dataframe
df = pd.DataFrame(data)
```

```python
# Exibo 'df'
df
```

```python
# Agrupo por empresas
group = df.groupby('Empresa')
```

```python
# Agora chamo a função 'sum' para saber o total de venda em cada empresa
group.sum(numeric_only=True)
```

```python
# Agora calculo a média de vendas em cada empresa
group.mean(numeric_only=True)
```

```python
# chamo a função 'describe' para gerar informações estatísticas de cada empresa
group.describe()
```

```python
# Agora agrupo por nomes
group = df.groupby('Nome')
```

```python
# Chamo a função 'sum' para buscar as vendas de cada pessoa
group.sum(numeric_only=True)
```

```python
# Aqui eu busco a venda total de uma pessoa em específico
group.sum(numeric_only=True).loc['Amy']
```

Video: Concatenar, Juntar e Mesclar

```python
# Crio o 'df1'
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
```

```python
# Crio o 'df2'
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7]) 
```

```python
# Crio o 'df3'
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
```

```python
# Exibo 'df1'
df1
```

```python
# Exibo 'df2'
df2
```

```python
# Exibo 'df3'
df3
```

```python
# Aqui eu junto os três dfs
pd.concat([df1,df2,df3])
```

```python
# Desta vez eu junto os dfs pelos indices
# se os indices nao coincidirem, o resultado vai ser NaN
pd.concat([df1,df2,df3],axis=1)
```

```python
# Crio o dataframe 'esquerda'
esquerda = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

# Crio o dataframe 'direita'
direita = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
# Exibo 'esquerda'
esquerda
```

```python
# Exibo 'direita'
direita
```

```python
# Junto os dois dataframes pela coluna 'key'
pd.merge(esquerda,direita,how='inner',on='key')
```

```python
# Crio dataframes com chaves compostas
```

```python

esquerda = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
```

```python
direita = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
esquerda
```

```python
direita
```

```python
# Junto os dois dfs onde 'key1' e 'key2' são iguais
pd.merge(esquerda, direita, on=['key1', 'key2'])
```

```python
# Junto os dois dfs por 'key1' e 'key2', mantendo todas as linhas de ambos dfs
# e onde não houver correspondência, o resultado é NaN
pd.merge(esquerda, direita, how='outer', on=['key1', 'key2'])
```

```python
# Mantenho todas as linhas do df da direita
# e onde não tiver correspondência na esquerda, o resultado é NaN
pd.merge(esquerda, direita, how='right', on=['key1', 'key2'])
```

```python
# Mantenho todas as linhas do df da esquerda
# e onde não tiver correspondência na direita, o resultado é NaN
pd.merge(esquerda, direita, how='left', on=['key1', 'key2'])
```

```python
# Crio dois dataframes
esquerda = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

direita = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
# Exibo 'esquerda'
esquerda
```

```python
# Exibo 'direita'
direita
```

```python
# Junto os dfs pelo indices
# mantém tudo o que está na esquerda, e procura na direita
esquerda.join(direita)
```

```python
# Junto os dois dfs, sem perder linhas de ambos
# e se o indice existir só em um df, aparece NaN no outro
esquerda.join(direita, how='outer')
```
