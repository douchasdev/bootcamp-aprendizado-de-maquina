** Importe pandas e leia o arquivo csv "Ecommerce Purchases" e configure-o para um DataFrame chamado ecom. **

```python
# Importa a biblioteca Pandas e atribui o apelido 'pd'
import pandas as pd
```

```python
# Carrega um arquivo CSV chamado 'Ecommerce Purchases' para um DataFrame do Pandas
ecom = pd.read_csv('Ecommerce Purchases')
```

** Quais foram os preços de compra mais altos e mais baixos? **

** Quantas pessoas têm Inglês 'en' como sua língua de escolha no site? **

```python
# Conta quantos registros possuem o idioma configurado como inglês ('en')
# 1. ecom['Language'] == 'en': Cria uma máscara de True/False para cada linha
# 2. ecom[...]: Aplica essa máscara para selecionar apenas as linhas True
# 3. len(...): Conta o número total de linhas resultantes desse filtro
len(ecom[ecom['Language'] == 'en'])
```

** Quantas pessoas têm o cargo de "Advogado"? **

```python
# Conta quantos clientes na base de dados possuem a profissão 'Lawyer' (Advogado)
# 1. ecom['Job'] == 'Lawyer': Filtra as linhas onde a coluna de profissão corresponde exatamente ao termo
# 2. len(...): Retorna a quantidade total de registros encontrados
len(ecom[ecom['Job'] == 'Lawyer'])
```

** Quantas pessoas fizeram a compra durante a AM e quantas pessoas fizeram a compra durante o PM? **

** (Sugestão: Confira [value_counts ()](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.value_counts.html)) **

```python
# Conta a frequência de cada valor único na coluna 'AM or PM'
ecom['AM or PM'].value_counts()
```

** Quais são os 5 títulos de trabalho mais comuns? **

```python
# Identifica as 5 profissões mais frequentes na base de dados
# 1. ecom['Job']: Seleciona a coluna de profissões
# 2. .value_counts(): Conta quantas vezes cada profissão aparece (do maior para o menor)
# 3. .head(): Filtra apenas as 5 primeiras do ranking (o "Top 5")
ecom['Job'].value_counts().head()
```

** Alguém fez uma compra que veio do Lot: "90 WT", qual foi o preço de compra para esta transação? **

```python
# Busca o preço de compra para um lote específico
# 1. ecom['Lot'] == '90 WT': Procura na coluna 'Lot' o valor exato '90 WT'
# 2. ecom[...]: Seleciona a(s) linha(s) que atendem a esse critério
# 3. ['Purchase Price']: Retorna apenas o valor da coluna de preço para esse resultado
ecom[ecom['Lot'] == '90 WT']['Purchase Price']
```

** Qual é o email da pessoa com o seguinte número do cartão de crédito: 4926535242672853 **

```python
# Localiza o endereço de e-mail associado a um número de cartão de crédito específico
# 1. ecom['Credit Card'] == 49265352...: Filtra a linha que possui esse número exato
# 2. ecom[...]: Seleciona essa linha específica no DataFrame
# 3. ['Email']: Extrai apenas a informação da coluna de e-mail
ecom[ecom['Credit Card'] == 4926535242672853]['Email']
```

** Quantas pessoas têm o American Express como seu fornecedor de cartão de crédito * e * fizeram uma compra acima de US $ 95? **

```python
# Conta quantas compras atendem a dois critérios simultâneos:
# 1. O cartão de crédito deve ser 'American Express'
# 2. O preço da compra deve ser maior que 95
len(ecom[(ecom['CC Provider'] == 'American Express') & (ecom['Purchase Price'] > 95)])
```

** Difícil: quantas pessoas tem um cartão de crédito que expira em 2025? **

```python
# Conta quantos cartões de crédito expiram no ano de '25'
# 1. .apply(lambda ano : ano[3:]): Percorre a coluna 'CC Exp Date' e extrai 
#    os caracteres do índice 3 em diante (ex: de '01/25' sobra apenas '25')
# 2. == '25': Compara se o resultado dessa extração é igual à string '25'
# 3. sum(...): Soma os valores True (que valem 1). O resultado é o total de cartões
sum(ecom['CC Exp Date'].apply(lambda ano : ano[3:]) == '25')
```

** Difícil: quais são os 5 principais provedores de e-mail / hosts mais populares (por exemplo, gmail.com, yahoo.com, etc ...) **

```python
# Identifica os 5 domínios de e-mail (provedores) mais comuns na base de dados
# 1. ecom['Email']: Seleciona a coluna de e-mails
# 2. .apply(lambda ...): Aplica uma função a cada e-mail da lista
# 3. dominio.split('@')[1]: Divide o e-mail no '@' e pega a segunda parte (o domínio)
# 4. .value_counts().head(): Conta a frequência dos domínios e mostra o Top 5
ecom['Email'].apply(lambda dominio : dominio.split('@')[1]).value_counts().head()
```
