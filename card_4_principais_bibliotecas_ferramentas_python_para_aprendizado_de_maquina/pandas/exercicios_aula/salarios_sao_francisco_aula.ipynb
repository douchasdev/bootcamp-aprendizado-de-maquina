** Leia o arquivo Salaries.csv como um DataFrame chamado "sal". **

```python
# Carrega o arquivo CSV 'Salaries.csv' e o armazena na variável 'sal'
sal = pd.read_csv('Salaries.csv')
```

** Qual é o "BasePay" médio? **

```python
# Calcula a média aritmética da coluna 'BasePay'
sal['BasePay'].mean()
```

** Qual é a maior quantidade de "OvertimePay" no conjunto de dados? **

```python
# Retorna o maior valor encontrado na coluna 'OvertimePay'
sal['OvertimePay'].max()
```

** Qual é o cargo de JOSEPH DRISCOLL? Nota: use todas as maiúsculas, caso contrário você pode obter uma resposta que não coincida (há também um Joseph Driscoll com minúsculas). **

```python
# Busca o cargo (JobTitle) do funcionário com o nome 'JOSEPH DRISCOLL'
# 1. sal['EmployeeName'] == 'JOSEPH DRISCOLL': Cria o filtro para o nome
# 2. sal[...]: Seleciona a linha correspondente a esse filtro
# 3. ['JobTitle']: Retorna apenas o conteúdo da coluna de cargo
sal[sal['EmployeeName'] == 'JOSEPH DRISCOLL']['JobTitle']
```

** Quanto JOSEPH DRISCOLL ganha (incluindo benefícios)? **

```python
# Retorna o valor total pago ao 'JOSEPH DRISCOLL', incluindo salário e benefícios
# 1. Filtra a linha onde o nome é exatamente 'JOSEPH DRISCOLL'
# 2. Seleciona a coluna 'TotalPayBenefits'
sal[sal['EmployeeName'] == 'JOSEPH DRISCOLL']['TotalPayBenefits']
```

** Qual o nome da pessoa mais bem paga (incluindo benefícios)? **

```python
# Localiza e exibe a linha inteira do funcionário com o maior 'TotalPayBenefits'
# 1. sal['TotalPayBenefits'].max(): Identifica qual é o maior valor numérico da coluna
# 2. sal['TotalPayBenefits'] == ...: Cria um boolean comparando cada linha com esse valor máximo
# 3. sal[...]: Extrai a linha onde essa comparação é verdadeira
sal[sal['TotalPayBenefits'] == sal['TotalPayBenefits'].max()]
```

** Qual o nome da pessoa paga mais baixa (incluindo benefícios)? Você percebe algo estranho sobre o quanto ele ou ela é paga? **

```python
# Localiza e exibe a linha inteira do funcionário com o menor 'TotalPayBenefits'
# 1. sal['TotalPayBenefits'].min(): Identifica o menor valor numérico da coluna
# 2. sal['TotalPayBenefits'] == ...: Filtra as linhas que possuem esse valor exato
sal[sal['TotalPayBenefits'] == sal['TotalPayBenefits'].min()]
```

** Qual foi a média (média) BasePay de todos os funcionários por ano? (2011-2014)? **

```python
# Calcula a média salarial base para cada ano registrado no dataset
# 1. .groupby('Year'): Agrupa todas as linhas que possuem o mesmo ano
# 2. .mean(numeric_only=True): Calcula a média de todas as colunas numéricas para cada grupo
# 3. ['BasePay']: Filtra o resultado para mostrar apenas a coluna de salário base
sal.groupby('Year').mean(numeric_only=True)['BasePay']
```

** Quantos títulos de trabalho únicos existem? **

```python
# Conta o número de cargos únicos no DataFrame
# 1. sal['JobTitle'].unique(): Retorna uma lista (array) com todos os nomes de cargos, sem repetições
# 2. len(...): Calcula o comprimento dessa lista, ou seja, a quantidade total de itens
len(sal['JobTitle'].unique())
```

** Quais são os 5 principais empregos mais comuns? **

```python
# Lista os 5 cargos mais frequentes no DataFrame 'sal'
# 1. .value_counts(): Conta quantas vezes cada cargo aparece e ordena do maior para o menor
# 2. .head(): Filtra apenas as 5 primeiras posições
sal['JobTitle'].value_counts().head()
```

** Quantos Títulos de Trabalho foram representados por apenas uma pessoa em 2013? (Por exemplo, títulos de trabalho com apenas uma ocorrência em 2013?) **

```python
# Conta quantos cargos tiveram apenas 1 ocorrência no ano de 2013
# 1. sal[sal['Year'] == 2013]: Filtra o banco de dados apenas para o ano de 2013
# 2. ['JobTitle'].value_counts(): Conta quantas pessoas existem em cada cargo
# 3. == 1: Cria uma lista de True/False, onde True significa que o cargo só aparece 1 vez
# 4. sum(...): Soma os valores True para chegar ao total final
sum(sal[sal['Year'] == 2013]['JobTitle'].value_counts() == 1)
```

** Quantas pessoas têm a palavra Chefe em seu cargo? (Isso é bastante complicado) **

```python
def cargo_chief(nome_cargo):
    # Converte o texto para minúsculo para garantir que 'Chief', 'CHIEF' ou 'chief' sejam detectados
    if 'chief' in nome_cargo.lower():
        return True
    else:
        return False
```

```python
# Conta quantos funcionários possuem a palavra 'chief' em seu cargo
# 1. .apply(lambda ...): Passa cada valor da coluna 'JobTitle' para a função 'cargo_chief'
# 2. cargo_chief(nome_cargo): Retorna True se 'chief' estiver no nome, False caso contrário
# 3. sum(...): Soma todos os True (1s) encontrados
sum(sal['JobTitle'].apply(lambda nome_cargo : cargo_chief(nome_cargo)))
```

** Bônus: Existe uma correlação entre o comprimento da seqüência do título do trabalho e o salário? **

```python
# Cria uma nova coluna chamada 'tamanho_do_titulo_do_trabalho'
# O valor de cada linha será a quantidade de caracteres do cargo (JobTitle)
sal['tamanho_do_titulo_do_trabalho'] = sal['JobTitle'].apply(len)
```

```python
# Gera uma matriz de correlação entre as duas colunas
# O resultado varia de -1 a 1
sal[['tamanho_do_titulo_do_trabalho', 'TotalPayBenefits']].corr()
```
