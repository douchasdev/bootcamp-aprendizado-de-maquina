```python
# chamo a biblioteca pandas e apelido de pd
import pandas as pd
```

```python
# leio o arquivo .csv obtido no kaggle https://www.kaggle.com/datasets/showmik121/smartphones-dataset-2026-1000-devices
celulares = pd.read_csv('smartprix_smartphones_april_2026.csv')
```

```python
# utilizo esta função do dataframe para saber quais informações me interessam
celulares.info()
```

```python
# selecionei as informações que mem interessam
celulares = celulares[['brand_name', 'model', 'price', 'has_5G', 'processor_brand', 'processor_name', 'ram', 'memory']]
```

```python
# Confirmo as alterações
celulares
```

```python
# renomeio a coluna 'price'
celulares = celulares.rename(columns={'price':'price (Indian Rupe)'})
```

```python
# confirmo a alteração
celulares
```

```python
# quantidade de celulares por marca
celulares['brand_name'].value_counts()
```

```python
# preço medio por marca
celulares.groupby('brand_name')['price (Indian Rupe)'].mean().sort_values(ascending=False)
```

```python
# quantidade de precessadores por marca
celulares['processor_brand'].value_counts()
```

```python
# média de preço por tipo de processador
celulares.groupby('processor_brand')['price (Indian Rupe)'].mean()
```

```python
# relação ram x preço
celulares[['ram', 'price (Indian Rupe)']].corr()
```

```python
# a ram tem uma influência moderada em relação ao preço
```

```python
# celulares que não possuem 5G
celulares[celulares['has_5G'] == False]
```

```python
# quantidade de celulares sem 5G
celulares_sem_5G = celulares[celulares['has_5G'] == False]
len(celulares_sem_5G)
```

```python
# processador que aparece mais vezes
celulares['processor_name'].value_counts().idxmax()
```

```python
# esta função lambda cria uma nova coluna com base na coluna de memória ram
celulares['nivel_ram'] = celulares['ram'].apply(
    lambda x: 'baixa' if x <= 4 else 'media' if x <= 8 else 'alta'
)
```

```python
# confirmo a alteração feita
celulares
```

```python
# exporta o dataframe celulares
celulares.to_csv('celulares_pratica.csv')
```
