3. Dado o array arr = np.array([10, 20, 30, 40, 50]), substitua todos os valores maiores que 30 por -1.

```python
# Cria um array com 5 elementos inteiros
arr = np.array([10, 20, 30, 40, 50])

# Aplica um boolea para filtragem e substituição:
# 1. arr > 30: Cria uma lista de True/False ([False, False, False, True, True]).
# 2. arr[...]: Seleciona apenas os elementos onde a condição é True (40 e 50).
# 3. = -1: Atribui o valor -1 a esses elementos selecionados.
arr[arr > 30] = -1
```

4. Operações Estatísticas: Crie uma matriz 5x5 de números inteiros aleatórios e calcule:

A média de cada coluna.
O valor máximo de cada linha.

```python
# Cria uma matriz 5x5 preenchida com números inteiros aleatórios entre 0 e 99.
matriz = np.random.randint(0, 100, (5, 5))

# Calcula a média aritmética de cada coluna, axis=0
# O resultado será um array com 5 valores (um para cada coluna).
media_colunas = matriz.mean(axis=0)

# Identifica o valor máximo de cada linha, axis=1
# O resultado será um array com 5 valores (o maior de cada uma das 5 linhas).
max_linhas = matriz.max(axis=1)
```

```python
# Exibe 'matriz'
matriz
```

```python
# Exibe 'media_colunas'
media_colunas
```

```python
# Exibe 'max_linhas'
max_linhas
```

Exercicios de pandas

```python
# Importa a biblioteca Pandas e a apelida como 'pd'.
import pandas as pd
```

```python
# Definindo os dados base
dados = {
    'Nome': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eduardo', 'Fernanda', 'Gabriel', 'Helena', 'Ítalo', 'Julia'],
    'Departamento': ['TI', 'RH', 'TI', 'Vendas', 'Financeiro', 'RH', 'Vendas', 'TI', 'Financeiro', 'TI'],
    'Salario': [8500, 4200, np.nan, 3500, 6000, np.nan, 3800, 10500, 5800, 7200],
    'Anos_Empresa': [5, 2, 8, 1, 4, 3, 2, 10, 5, 6]
}
```

```python
# Criando o DataFrame
df = pd.DataFrame(dados)
```

```python
# Exibe 'df'
df
```

5. Selecione apenas os funcionários do departamento 'TI' que ganham mais de R$ 5.000

```python
# 1. (df['Departamento'] == 'TI'): Cria um boolean para quem é do setor de TI.
# 2. (df['Salario'] > 5000): Cria um boolean para salários maiores que 5000.
# 3. &: O operador 'E' lógico exige que AMBAS as condições sejam verdadeiras.
# 4. df[...]: Aplica o filtro final, retornando apenas as linhas que atendem aos critérios.
ti_bem_pago = df[(df['Departamento'] == 'TI') & (df['Salario'] > 5000)]
```

```python
# Exibe 'ti_bem_pago'
ti_bem_pago
```

6. Calcule a média salarial por departamento

```python
# 1. .groupby('Departamento'): Divide o DataFrame em grupos baseados nos valores da coluna 'Departamento'.
# 2. ['Salario']: Seleciona apenas a coluna de salários dentro de cada grupo para o cálculo.
# 3. .mean(): Aplica a função de média aritmética para cada um desses grupos.
media_departamento = df.groupby('Departamento')['Salario'].mean()
```

```python
# Exibe 'media_departamento
media_departamento
```

7. Identifique se existem valores nulos no DataFrame. Substitua os valores nulos da coluna Salario pela mediana dos salários

```python
# Identificar os nulos
nulos = df.isna().sum()
```

```python
# Exibe 'nulos'
nulos
```

```python
# Substituir pela mediana:
mediana = df['Salario'].median()

# Preenche os valores nulos (vazios) da coluna 'Salario' com a mediana calculada.
# O método .fillna() substitui entradas NaN pelo valor fornecido 
# e o resultado é atribuído de volta à coluna original para salvar as alterações.
df['Salario'] = df['Salario'].fillna(mediana)

# Exibe 'df'
df
```

8. Crie uma nova coluna chamada Bonus, que corresponde a 10% do salário para quem tem mais de 5 anos de empresa e 5% para os demais

```python
# Cria uma nova coluna 'Bonus' baseada no tempo de empresa do funcionário.
# np.where(condição, valor_se_verdadeiro, valor_se_falso):
# 1. df['Anos_Empresa'] > 5: A condição a ser testada (mais de 5 anos de empresa).
# 2. df['Salario'] * 0.10: O que acontece se for True (bônus de 10%).
# 3. df['Salario'] * 0.05: O que acontece se for False (bônus de 5%).
df['Bonus'] = np.where(df['Anos_Empresa'] > 5, df['Salario'] * 0.10, df['Salario'] * 0.05)
```

```python
# Exibe 'df'
df
```

9. Mostre os 3 funcionários mais bem pagos da empresa.

```python
# 1. .sort_values(by='Salario'): Define que a ordenação será baseada na coluna 'Salario'.
# 2. ascending=False: Define a ordem decrescente (do maior para o menor). 
# 3. .head(3): Após ordenar, seleciona apenas as 3 primeiras linhas do resultado.
df.sort_values(by='Salario', ascending=False).head(3)
```

Exercicios de numpy e pandas

10. Crie uma matriz NumPy 10x4 com números aleatórios simulando notas de 4 provas para 10 alunos

```python
# Gera floats entre 0 e 10 para 10 linhas e 4 colunas
notas = np.random.uniform(0, 10, (10, 4))
notas
```

11. Converta essa matriz em um DataFrame do Pandas

```python
# Definimos nomes para as colunas
provas = ['Prova_1', 'Prova_2', 'Prova_3', 'Prova_4']
df_alunos = pd.DataFrame(notas, columns=provas)

# Exibe 'df_alunos'
df_alunos
```

12. Adicione uma coluna chamada Media calculando a média das notas de cada aluno

```python
# Coluna chamada "Media" calculando a média das notas de cada aluno
# axis=1 indica que queremos a média na horizontal (por linha/aluno)
df_alunos['Media'] = df_alunos.mean(axis=1)

# Exibe 'df_alunos'
df_alunos
```

13. Adicione uma coluna Status que diz "Aprovado" se a média for maior que 7 e "Reprovado" caso contrário

```python
# Coluna "Status" (Aprovado >= 7 / Reprovado < 7)
# np.where(condição, se_verdadeiro, se_falso)
df_alunos['Status'] = np.where(df_alunos['Media'] >= 7, 'Aprovado', 'Reprovado')
```

```python
# Exibe 'df_alunos'
df_alunos
```
