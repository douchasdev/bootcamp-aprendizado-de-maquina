# Exercícios NumPy  

Agora que aprendemos sobre NumPy, vamos testar seu conhecimento. Começaremos com algumas tarefas simples, para depois entrarmos nas perguntas mais complicadas.

#### Importe NumPy como np

```python
# Importa a biblioteca NumPy, utilizando o apelido padrão 'np'
import numpy as np
```

#### Crie uma matriz de 10 zeros

```python
# Cria um array unidimensional (vetor) com 10 elementos, todos inicializados com o valor 0
np.zeros(10)
```

#### Crie uma matriz de 10 ones

```python
# Cria um array com 10 elementos, todos inicializados com o valor 1
np.ones(10)
```

#### Crie uma matriz de 10 cincos

```python
# Cria um array com 10 elementos igualmente espaçados entre o início (5) e o fim (5)
np.linspace(5, 5, 10)
```

#### Crie um array de inteiros de 10 até 50

```python
# Gera um array com uma sequência de números inteiros começando em 10 e indo até 50
np.arange(10, 51)
```

#### Crie um array dos numeros pares de 10 até 50

```python
# Gera um array começando em 10 e indo até 50, pulando de 2 em 2
np.arange(10, 51, 2)
```

#### Criei uma matriz 3x3 com valores variando de 0 até 8

```python
# Cria um array unidimensional (vetor) com 9 elementos, de 0 a 8
arr = np.arange(0, 9)

# Reorganiza os dados do array 'arr' em uma nova forma (shape) de 3 linhas e 3 colunas
matriz = arr.reshape((3,3))

# Exibe a matriz
matriz
```

#### Crie uma matriz identidade 3x3

```python
# Cria uma matriz identidade de ordem 3x3
np.eye(3)
```

#### Use NumPy para gerar números aleatórios entre 0 e 1

```python
# Gera um array contendo 1 único número aleatório
# O número é extraído de uma distribuição uniforme no intervalo [0, 1)
np.random.rand(1)
```

#### Use Numpy para gerar um array de 25 números aleatórios tirados de uma distribuição normal.

```python
# Gera um array com 25 números aleatórios extraídos de uma "Distribuição Normal Padrão"
np.random.randn(25)
```

#### Crie a seguinte matriz:

```python
# Cria uma matriz 10x10 com valores que variam de 0.01 a 1.0
np.arange(1, 101).reshape((10, 10)) / 100
```

#### Crie um array de tamanho 20 igualmente espaçado entre 0 e 1.

```python
# Cria um array com 20 números igualmente espaçados, começando em 0 e terminando em 1
np.linspace(0, 1, 20)
```

## Indexação Numpy e Seleção

Agora você receberá algumas matrizes e será solicitado a replicar as saídas resultantes da matriz:

```python
# Cria uma matriz 5x5 com valores sequenciais de 1 a 25
mat = np.arange(1,26).reshape(5,5)

# Exibe a matriz
mat
```

```python
# '2:' seleciona da linha de índice 2 até o final (linhas 3, 4 e 5)
# '1:' seleciona da coluna de índice 1 até o final (colunas 2, 3, 4 e 5)
mat[2: , 1: ]
```

```python
# Acessa o elemento localizado na linha de índice 3 e coluna de índice 4
mat[3, 4]
```

```python
# Extrai uma submatriz das 3 primeiras linhas (índices 0, 1 e 2)
# e da segunda coluna (índice 1)
mat[0:3, 1:2]
```

```python
# Seleciona a última linha da matriz (índice 4)
mat[4, 0: ]
```

```python
# Seleciona todas as linhas a partir do índice 3 até o final
# isso resultará nas duas últimas linhas completas
mat[3: ,0: ]
```
