```python
# Cria um array de 5 elementos contendo números aleatórios
# segue distribuição uniforme (todos os números têm a mesma chance de aparecer)
np.random.rand(5)
```

```python
# Gera 5 números aleatórios (0 a 1), escala para o intervalo (0 a 100), 
# e arredonda o resultado para o número inteiro mais próximo com 0 casas decimais.
np.round(np.random.rand(5) * 100, 0)
```

```python
# Cria um array unidimensional de 25 elementos
# preenchido com números aleatórios entre 0.0 e 1.0
arr = np.random.rand(25)
```

```python
# Exibe 'arr'
arr
```

```python
# Retorna uma tupla contendo as dimensões do array 'arr'
arr.shape
```

```python
# Retorna o valor máximo presente em toda a matriz 'arr'
arr.max()
```

```python
# Retorna o valor mínimo presente em toda a matriz 'arr'
arr.min()
```

```python
# Retorna o índice da posição do maior valor encontrado no array
arr.argmax()
```

```python
# Retorna o índice da posição do menor valor encontrado no array
arr.argmin()
```

Video: Indexacao e Fatiamento de arrays

```python
# Cria um array com valores de 0 até (mas não incluindo) 30,
# incrementando de 3 em 3.
arr = np.arange(0, 30, 3)
```

```python
# Exibe 'arr'
arr
```

```python
# Acessa o elemento na quarta posição do array 'arr'
arr[3]
```

```python
# Retorna os elementos do índice 2 até o índice 4 (o índice 5 não é incluído)
arr[2:5]
```

```python
# O resultado são os 5 primeiros elementos do array
arr[0:5]
```

```python
# Retorna do índice 2 até o final do array.
arr[2:]
```

```python
# Modifica o array 'arr' atribuindo o valor 100 
# a todos os elementos a partir do índice 2 até o final
arr[2:] = 100
```

```python
# Exibe 'arr'
arr
```

```python
# Cria um array com 50 números (de 0 a 49) 
# e o organiza em uma matriz de 5 linhas e 10 colunas
arr = np.arange(50).reshape((5, 10))
```

```python
# Exibe 'arr'
arr
```

```python
# Retorna uma tupla contendo as dimensões do array 'arr'
arr.shape
```

```python
# Cria uma cópia independente das 3 primeiras linhas de 'arr'
# O resultado 'arr2' será uma matriz de 3 linhas e 10 colunas (shape: (3, 10))
arr2 = arr[:3][:].copy()
```

```python
# Exibe 'arr2'
arr2
```

```python
# Atribui o valor 100 a todos os elementos da matriz 'arr2'
# Como 'arr2' é uma cópia independente, o 'arr' original não é afetado
arr2[:] = 100
```

```python
# Exibe 'arr2'
arr2
```

```python
# Exibe 'arr'
arr
```

```python
# Seleciona as linhas de 1 a 3 e as colunas de 5 até a última
arr[1:4, 5:]
```

```python
# O resultado será True para cada célula onde o valor for maior que 50, e False caso contrário.
bool = arr > 50
```

```python
# Retornando apenas os elementos 
# onde o valor booleano correspondente é True.
arr[bool]
```

```python
# Cria um array com 30 números igualmente espaçados 
# no intervalo fechado de 0 a 100.
array = np.linspace(0, 100, 30)
```

```python
# Exibe 'array'
array
```

```python
# Retorna uma tupla contendo as dimensões do 'array'
array.shape
```

```python
# Reorganiza o array 'array' (que tinha 30 elementos) em uma nova estrutura
# Agora ele se torna uma matriz de 2 dimensões com 3 linhas e 10 colunas
array = array.reshape(3, 10)
```

```python
# Exibe 'array'
array
```

```python
# Retorna uma tupla contendo as dimensões do array 'array'
array.shape
```

```python
# Acessa o elemento na linha 0, coluna 2 e o elemento na linha 1, coluna 2
array[0:2, 2]
```

Video: Operacoes com Numpy Arrays

```python
# Cria um array unidimensional com números inteiros de 0 até 15.
arr = np.arange(0, 16)
```

```python
# Exibe 'arr'
arr
```

```python
# Realiza a soma elemento a elemento
arr + arr
```

```python
# Realiza a subtração elemento a elemento
arr - arr
```

```python
# Realiza a multiplicação elemento a elemento
arr * arr
```

```python
# Realiza a divisão elemento a elemento
arr / arr
```

```python
# Divide um por zero
1 / 0
```

```python
# Realiza a divisão de 1 por cada elemento do array 'arr'.
1 / arr
```

```python
# Eleva cada elemento do array 'arr' ao quadrado
arr ** 2
```

```python
# Adiciona o valor 100 a cada elemento do array 'arr'
arr + 100
```

```python
# Subtrai o valor 100 a cada elemento do array 'arr'
arr - 100
```

```python
# Multiplica o valor 100 a cada elemento do array 'arr'
arr * 100
```

```python
# Calcula a raiz quadrada de cada elemento do array 'arr'
np.sqrt(arr)
```

```python
# Calcula a exponencial (e^x) de cada elemento do array 'arr'
np.exp(arr)
```

```python
# Calcula a média aritmética de todos os elementos do array 'arr'
np.mean(arr)
```

```python
# Calcula o desvio padrão (standard deviation) dos elementos do array 'arr'
np.std(arr)
```

```python
# Calcula o seno de cada elemento do array 'arr'
np.sin(arr)
```

```python
# Retorna o maior valor contido no array 'arr'
np.max(arr)
```

```python
# Retorna o menor valor contido no array 'arr'
np.min(arr)
```

```python
# É o método equivalente à função np.max(arr)
arr.max()
```

```python
# É o método equivalente à função np.min(arr)
arr.min()
```
