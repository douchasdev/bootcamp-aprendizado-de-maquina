Video: Indexacao e Fatiamento de arrays

```python
# cria um array com elementos de 0 a 29, indo de 3 em 3 
arr = np.arange(0, 30, 3)
```

```python
# rxibe 'arr'
arr
```

```python
# acessa o elemento na quarta posição do array arr
arr[3]
```

```python
# retorna os elementos presentes nos indices 2,3 e 4
arr[2:5]
```

```python
# retorna os elementos presentes nos indices 0,1,2,3 e 4
arr[0:5]
```

```python
# retorna os elementos presentes nos indices 2,3,4,...,9 (final)
arr[2:]
```

```python
# todos os elementos a partir do indice 2 até final são 100
arr[2:] = 100
```

```python
# Exibe 'arr'
arr
```

```python
# cria um array de 0 até 49 e o transforma em uma matriz 5x10
arr = np.arange(50).reshape((5, 10))
```

```python
# exibe 'arr'
arr
```

```python
# retorna as dimensões da matriz arr
arr.shape
```

```python
# cria uma cópia da matriz arr a partir das suas 3 primeiras linhas
arr2 = arr[:3][:].copy()
```

```python
# exibe 'arr2'
arr2
```

```python
# todos os elementos de arr2 são 100
arr2[:] = 100
```

```python
# exibe 'arr2'
arr2
```

```python
# exibe 'arr'
arr
```

```python
# seleciona as linhas de indices 1,2 e 3 e as colunas de indices 5,6,7,8 e 9
arr[1:4, 5:]
```

```python
# array de True ou False, o indice em que o elemento for maior que 50 retorna True, se for menor retorna False
bool = arr > 50
```

```python
# retorna um array em que os elementos são maiores que 50
arr[bool]
```

Video: Operacoes com Numpy Arrays

```python
# cria um array de uma dimensão com elementos indo de 0 até 15
arr = np.arange(0, 16)
```

```python
# Exibe arr
arr
```

```python
# realiza a soma elemento a elemento
arr + arr
```

```python
# realiza a subtração elemento a elemento
arr - arr
```

```python
# realiza a multiplicação elemento a elemento
arr * arr
```

```python
# realiza a divisão elemento a elemento
arr / arr
```

```python
# divide um por zero
1 / 0
```

```python
# realiza a divisão de 1 por cada elemento do array arr
1 / arr
```

```python
# eleva cada elemento do array arr ao quadrado
arr ** 2
```

```python
# adiciona o valor 100 a cada elemento do array arr
arr + 100
```

```python
# subtrai o valor 100 a cada elemento do array arr
arr - 100
```

```python
# multiplica o valor 100 a cada elemento do array arr
arr * 100
```

```python
# calcula a raiz quadrada de cada elemento do array arr
np.sqrt(arr)
```

```python
# calcula a exponencial de cada elemento do array arr
np.exp(arr)
```

```python
# calcula a média aritmética de todos os elementos do array arr
np.mean(arr)
```

```python
# calcula o desvio padrão do array arr
np.std(arr)
```

```python
# calcula o seno de cada elemento do array arr
np.sin(arr)
```

```python
# retorna o maior valor contido no array arr
np.max(arr)
```

```python
# retorna o menor valor contido no array arr
np.min(arr)
```

```python
# é o método equivalente à função np.max(arr)
arr.max()
```

```python
# é o método equivalente à função np.min(arr)
arr.min()
```
